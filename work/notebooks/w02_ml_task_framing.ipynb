## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed engineering rule limits human prioritization because it weights search inputs linearly and uniformly across all clients. For instance, a manual rule might state that any page with over 180 days since its last update and over 500 impressions is automatically a "refresh candidate".  However, real-world search spaces are non-linear; different client profiles exhibit highly unbalanced traffic baselines and seasonal noise patterns. Machine learning models (such as Random Forests or Decision Trees) beat static rules because they evaluate multi-dimensional correlations simultaneously—such as structural interactions between click-through rates, position tier decay, and user scroll engagement—ranking the high-value opportunities with a massive lift in precision.
